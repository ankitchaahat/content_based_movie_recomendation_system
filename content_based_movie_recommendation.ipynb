```python
import numpy as np
import pandas  as pd
```

```python
movies=pd.read_csv('/content/tmdb_5000_movies.csv')
credits=pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)['crew']
```

```python
# merging the dataset
movies=movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies.info()
```

```python
# these are the columns we are going to use \
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# checking null values
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
# check duplicatd values
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# A function to acces the name key form the dictonary

# The ast module in Python is used to work with Abstract Syntax Trees, which is necessary for safely evaluating the string representation of your 'genres' column as Python lists.
import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# access the cast

def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])

  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# convert the overview column into a list
movies['overview']
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# remove the space between the two words
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
# new dataframe

new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# convert tags column into string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
# convert into lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
# stemming
# from nltk.stem.porter import PorterStemmer
# ps=PorterStemmer()

def stem(text):
  y=[]
  for i in text.split():
    y.append(nltk.PorterStemmer().stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]  # stemmed
```

```python
# data transformation  converting into vectors and removing stop words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
print(vectors)
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()[1:500]
```

```python
# using cosine similarity to find the similarities among movies
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors)   # passing all the vectors at once
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0] # similarity of 1st movies with each movie (1 shows the similary with itself )
```

```python
similarity[0].shape
```

```python
new_df.head()
```

```python
# sorting the list on the basis of vector similarity not index based
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]]['title'])
```

```python
recommend('Batman')
```
